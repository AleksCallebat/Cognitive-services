```python
from difflib import SequenceMatcher
from mstranslator import Translator
translator = Translator("9c5218a025bf47728bf04cacca30583f")
from mediawiki import MediaWiki

def ecart_traduction(page="Paris",lan1="en",lan2="fr"):
    wikipedia = MediaWiki(lang=lan1)
    version1 = wikipedia.page(page)
    text1=version1.content
    sections1=version1.sections
    text1=[text1.section(k) for k in sections1]
    
    wikipedia = MediaWiki(lang=lan2)
    wikipedia.set_lang(lan2)
    version2 = wikipedia.page(page)
    text2=version2.content
    sections2=text2.sections
    sections2bis=[translator.translate(s,lang_from=lan1,lang_to=lan2) for s in sections2]
    text2=[translator.translate(text2.section(s), lang_from=lan1, lang_to=lan2) for k in sections2]
    

    return([SequenceMatcher(None, text1[0], text2[0]).ratio(),text1,text2])
```

```python
print(ecart_traduction("Paris")[0])
```
